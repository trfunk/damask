```python
import damask
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
grains_20 = pd.read_csv("20_32x32x32_pc_vor_tension2.5.sta", delim_whitespace=True)
grains_200 = pd.read_csv("200_32x32x32_pc_vor_tension2.5.sta", delim_whitespace=True)
grains_2000 = pd.read_csv("2000_32x32x32_pc_vor_tension2.5.sta", delim_whitespace=True)
grains_20.head()
grains_200.head()
grains_2000.head()
```

```python
plt.figure(dpi = 500) 
sns.lineplot(data=grains_20, x="Increment", y="IterationsNeeded", errorbar=None)
sns.lineplot(data=grains_200, x="Increment", y="IterationsNeeded", errorbar=None)
sns.lineplot(data=grains_2000, x="Increment", y="IterationsNeeded", errorbar=None)
plt.legend(["20 Körner", "200 Körner", "2000 Körner"])
plt.xlabel("Inkrement")
plt.ylabel("Iterationen")
plt.title("Benötigte Iterationen pro Inkrement")

plt.savefig("increments.pdf")
```
